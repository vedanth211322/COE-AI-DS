**Drop rows with missing values**

```python
uber_data=uber_data.dropna()
uber_data
```

**fill missing values (propose column with unknown value)**

```python
uber_data.loc[:, 'CATEGORY*'] = uber_data['CATEGORY*'].fillna('Unknown')
uber_data.loc[:,'STOP*'] = uber_data['STOP*'].fillna('Unknown Location')
uber_data.loc[:,'PURPOSE*'] = uber_data['PURPOSE*'].fillna('Unknown Purpose')
uber_data
```

```python
uber_data.isnull().sum()
```

**Check and remove duplicates**

```python
unique_dates = uber_data['START_DATE*'].unique()
unique_dates
```

```python
duplicate_dates = uber_data['START_DATE*'].duplicated().sum() #.duplicated checks
duplicate_dates
```

```python
uber_data = uber_data.drop_duplicates() #drop duplicates
uber_data
```

```python
duplicates = uber_data[uber_data.duplicated()].sum()
duplicates
```

**Convert START_DATE and END_DATE to datetime**

```python
uber_data.loc[:,'START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'],errors='coerce')
uber_data.dtypes
```

```python
uber_data.loc[:, 'END_DATE*'] = pd.to_datetime(uber_data['END_DATE*'], errors='coerce')
uber_data.dtypes
```

**Total number of rides per category:**

```python
rides = uber_data.groupby('CATEGORY*').size()
rides
```

**Total miles traveled for each purpose:**

```python
miles_travelled = uber_data.groupby('PURPOSE*')['MILES*'].sum()
miles_travelled
```

**Average distance for business vs. personal rides:**

```python
business_miles = uber_data.groupby('CATEGORY*')['MILES*'].agg(['sum','mean','max'])
business_miles
```

**Add a column for cost estimation (assuming $2 per mile):**

```python
import numpy as np
#uber_data['cost']= np.where(uber_data['MILES*'] > 0,uber_data['MILES*']*2,0) getting warning
uber_data.loc[:, 'cost'] = np.where(uber_data['MILES*'] > 0, uber_data['MILES*'] * 2, 0)
uber_data.head(10)
```

**Filter rides longer than 50 miles:**

```python
longer_miles = uber_data[uber_data['MILES*'] >50]
longer_miles
```

**Filter by specific purpose (e.g., meetings):**

```python
purpose_filter = uber_data[uber_data['PURPOSE*'] == "Meeting"]
purpose_filter
```

**What is the total number of business trips versus personal trips?**

```python
business_trip_count = uber_data[uber_data['CATEGORY*'] == 'Business'].shape
business_trip_count
```

```python
personal_trip_count = uber_data[uber_data['CATEGORY*'] == 'Personal'].shape
personal_trip_count
```

**What percentage of trips are business versus personal?**

```python
category_counts = uber_data['CATEGORY*'].value_counts()

total_trips = category_counts.sum()

business_percentage = (category_counts.get('Business', 0) / total_trips) * 100
business_percentage
```

```python
personal_percentage = (category_counts.get('Personal', 0) / total_trips) * 100
personal_percentage
```
