```python
df['dt'] = pd.to_datetime(df['dt'], errors='coerce')
df['Year']=df['dt'].dt.year

avg_temp=df.groupby('Year')['AverageTemperature'].mean().reset_index()
plt.plot(avg_temp['Year'],avg_temp['AverageTemperature'],color='black')
plt.xlabel('year')
plt.ylabel('AverageTemperature')
plt.title('year vs AverageTemperature')
```

```python
avg_temp = df.groupby('Country')['AverageTemperature'].mean().reset_index()
avg_temp = avg_temp.dropna()
top = avg_temp.sort_values(by='AverageTemperature',ascending=False).head(10)
plt.figure(figsize=((12,6)))
plt.bar(top['Country'],top['AverageTemperature'],color='orange')
plt.title('Top 10 countries with highest avg temp',fontsize=10)
plt.xlabel('Country')
plt.ylabel('Avg Temp')
```

```python
#spread fo temp for a specific country over a given period
#x years or countries
#avg temp
#use boxplot

import matplotlib.pyplot as plt
grouped = df[df['Country'] == 'Denmark'].groupby('Year')['AverageTemperature'].apply(list)
plt.figure(figsize=(10, 6))
plt.boxplot(grouped)
plt.title('Spread of Average Temperature in Denmark Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.show()
```
