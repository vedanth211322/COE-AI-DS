```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y,color='r',marker='o',linestyle='dashed')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x=['jan','feb','mar','apr','may','jun']
y=[2000,3000,1500,4000,3000,2000]
plt.plot(x,y,color='green',linestyle='dashed',marker='o')
plt.xlabel('months')
plt.ylabel('savings')
plt.title('month vs savings')
```

```python
import matplotlib.pyplot as plt

x = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
y = [2000, 3000, 1500, 4000, 3000, 2000]

colors = ["orange", "skyblue", "brown", "yellow", "lightblue", "pink"]

plt.bar(x, y, color=colors)
plt.xlabel('Months')
plt.ylabel('Savings')
plt.title('Month vs Savings')
plt.show()
```

```python
import matplotlib.pyplot as plt

y = [2000, 3000, 1500, 3500,2000,2500,1500,1500]

plt.hist( y,bins=5, color='skyblue')
plt.xlabel('Months')
plt.ylabel('Savings')
plt.title('Month vs Savings')
plt.show()
```
