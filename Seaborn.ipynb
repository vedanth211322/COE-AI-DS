```python
import seaborn as sns
import pandas as pd
tips = sns.load_dataset("tips")
tips.head()
```

```python
#histogram
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.histplot(tips['total_bill'],bins=20,color='b',kde=True) #kde is another plot 
plt.title('distribution of total bills')
plt.show()
```

```python
sns.kdeplot(tips['total_bill'],fill=True,color='brown')
plt.show()
```

```python
#mean of a numeric variable (y)
sns.barplot(y=tips['total_bill'],x=tips['sex'],palette='pastel')
```

```python
sns.boxplot(y=tips['total_bill'],x=tips['sex'],palette='pastel')
```

```python
#The hue parameter in Seaborn is used to color the data points based on the values of a categorical variable.
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='deep')
```

```python
sns.lineplot(data=tips,x='sex',y='tip',marker='o',color='r')
```

```python
#It creates a grid of scatter plots (for numeric data) to show pairwise relationships between features
sns.pairplot(data=tips,hue='sex',palette='deep')
```

```python
sns.pairplot(data=tips, hue='sex', vars=['total_bill', 'tip', 'size'], palette='deep')
```

```python
sns.pairplot(data=tips, hue='sex', palette='deep', kind='reg')
```

```python
sns.pairplot(data=tips, hue='sex', palette='deep', markers=["o", "s"])
```

```python
# visualize correlation matrices, data tables, or any kind of two-dimensional data.
# Calculate the correlation matrix for numerical features
correlation_matrix = tips.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix of Tips Dataset')
plt.show()
```

```python
tips = sns.load_dataset('tips')
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_tips.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Tips Dataset')
plt.show()
```
